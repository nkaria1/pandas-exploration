```python
def circumference(half_circumference):
    circum=half_circumference*2
    return circum
df['circumference']=df['half circumference'].apply(circumference)
```

```python
df['class']=np.where(df['petal length']>3.2 ,  1, 0)
```

```python
age=[]

for row in sentosa['half circumference']:
    if (row <1.5):
        age.append('infant')
    elif (row >=1.5 and row<1.95):
        age.append('mature')
    else:
        age.append('adult')

sentosa['age']=age

sentosa.sort_values('half circumference', axis=0, ascending=True)
```
